## Some artificial data

```python
parent_income = np.random.normal(100, 1, size=100)
sns.displot(parent_income) ;
```

```python
df = pd.DataFrame(
    dict(height=np.random.normal(180, 1, size=100) + parent_income,
         weight=np.random.normal(80, 1, size=100) + parent_income,
         age=np.random.randint(8, 12, size=100),
         sex=pd.Categorical(np.random.randint(1, 3, size=100))
        )
)
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df.height, df.weight) ;
```

```python
sns.scatterplot(data=df, x='height', y='weight', hue='age') ;
```

```python
sns.scatterplot(data=df, x='height', y='weight', hue='sex') ;
```

```python
sns.displot(data=df, x='height', hue='sex') ;
```

```python
g = sns.FacetGrid(data=df, col='sex')
g.map(sns.histplot, 'height') ;            
```

```python
g = sns.FacetGrid(data=df, col='sex')
g.map(sns.scatterplot, 'height', 'weight', 'age') ;
plt.legend() ; 
```

```python
sns.lmplot(data=df, x='height', y='weight') ;
```
